```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from rdkit import Chem
from rdkit.Chem import Descriptors

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Thesis_Implementation/Dataset/test/forranking2.csv')

smiles = df['SMILES']
mols = [Chem.MolFromSmiles(x) for x in smiles]
df['MolWeight'] = [Descriptors.MolWt(x) for x in mols]
df['LogP'] = [Descriptors.MolLogP(x) for x in mols]
df['HydrogenBondDonors'] = [Descriptors.NumHDonors(x) for x in mols]
df['HydrogenBondAcceptors'] = [Descriptors.NumHAcceptors(x) for x in mols]

train_data = df.sample(frac=0.8, random_state=0)
test_data = df.drop(train_data.index)

X_train = train_data[['MolWeight', 'LogP', 'HydrogenBondDonors', 'HydrogenBondAcceptors']]
y_train = train_data['Affinity']

model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train, y_train)

X_test = test_data[['MolWeight', 'LogP', 'HydrogenBondDonors', 'HydrogenBondAcceptors']]
y_test = test_data['Affinity']

y_pred = model.predict(X_test)


error = np.abs(y_pred - y_test)
mean_error = np.mean(error)
print("Mean absolute error:", mean_error)

df['PredictedAffinity'] = model.predict(df[['MolWeight', 'LogP', 'HydrogenBondDonors', 'HydrogenBondAcceptors']])

df = df.sort_values(by=['MolWeight', 'PredictedAffinity'], ascending=[True, True])
```
